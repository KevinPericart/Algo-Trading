# <center> Bollinger Band application to NIO Stock </center>

```python
import pandas as pd
import plotly.graph_objs as go
from pandas_datareader import data, wb
import datetime
```

```python
start = pd.to_datetime('2018-01-01')
end = pd.to_datetime('2020-10-20')
```

```python
nio_df = data.DataReader('NIO', 'yahoo', start, end)
```

```python
nio_df
```

```python
nio_df['Middle Band'] = nio_df['Close'].rolling(window = 20).mean()
nio_df['Lower Band'] = nio_df['Middle Band'] - 1.96 * nio_df['Close'].rolling(window = 20).std()
nio_df['Upper Band'] = nio_df['Middle Band'] + 1.96 * nio_df['Close'].rolling(window = 20).std()
```

```python
nio_df
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = nio_df.index, y = nio_df['Middle Band'],line = dict(color = 'blue', width = .7), name = 'Middle Band'))
fig.add_trace(go.Scatter(x = nio_df.index, y = nio_df['Upper Band'],line = dict(color = 'red', width = 1.5), name = 'Upper Band (Sell)'))
fig.add_trace(go.Scatter(x = nio_df.index, y = nio_df['Lower Band'],line = dict(color = 'green', width = 1.5), name = 'Lower Band (Buy)'))
fig.add_trace(go.Candlestick(x = nio_df.index,open = nio_df['Open'], high = nio_df['High'], low = nio_df['Low'], close = nio_df['Close'], name = 'Market Data'))
fig.update_layout(title = 'Bollinger Band Strategy', yaxis_title = 'Nio Electric cars Stock Price (USD per Shares)')
fig.update_xaxes(rangeslider_visible = True,rangeselector = dict(buttons = list([dict(count = 1, label = "1M", step = "month", stepmode = "backward"), dict(count = 6, label = "6M", step = "month", stepmode = "backward"), dict(count = 1, label = "YTD", step = "year", stepmode = "todate"), dict(count = 1, label = "1Y", step = "year", stepmode="backward"), dict(step="all")])))
fig.show()
```
